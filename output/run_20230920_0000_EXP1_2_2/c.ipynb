# Vertical integration of carbon

```python
import vstats
```

```python
array1d_iT_C0to354  =vstats.vintegrate(array2d_idepth_iT_C, drF,  0, 73)
array1d_iT_C354to360=vstats.vintegrate(array2d_idepth_iT_C, drF, 73, 74)
array1d_iT_C0to360  =vstats.vintegrate(array2d_idepth_iT_C, drF,  0, 74)
```

# Annual drift of $C$

```python
nbdepth=RF.size-1
array2d_idepth_iyr2tolast_deltaC=np.zeros((nbdepth,nbyr-1))
for iyr in range(2,nbyr+1):
    iTcurrent=365*(iyr-1)+364
    array1d_idepth_Ccurrent=array2d_idepth_iT_C[:,iTcurrent]
    iTprevious=365*(iyr-2)+364
    array1d_idepth_Cprevious=array2d_idepth_iT_C[:,iTprevious]
    array1d_idepth_deltaC=array1d_idepth_Ccurrent-array1d_idepth_Cprevious
    array2d_idepth_iyr2tolast_deltaC[:,iyr-2]=array1d_idepth_deltaC
array1d_iyr_yr=np.arange(1,nbyr+1)
```
